```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.random.rand(3,3)
array3 = np.zeros((3,3))
```

```python
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)
```

```python
loaded_array1 = np.load('array1.npy')
print(loaded_array1)
```

```python
try:
    logo = np.load('numpy-logo.npy')
#now code to display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    logo = logo/255.0 #this is a normalization step if logo file contain pixel outside
    # the range[0,1] e.g, [0,255] for grayscale images then to normalize the values
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.grid(False)


except FileNotFoundError:
    print("numpy logo file not found")


```
