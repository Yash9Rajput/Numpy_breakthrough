```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau: ", sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data, axis = 0)) #if axis=0 it sum columnwise and if axis=1 it sum rowwise
#yearly_total= np.sum(sales_data[:,1:], axis = 0)
yearly_total= np.sum(sales_data[:,1:])
print("yearly_total: ",yearly_total)


#Mininum sales per restaurant
min_sales = np.min(sales_data[:,1:] ,axis = 1) #axis=1 is to search rowwise
print("min_sales: ", min_sales)

#Maximum sales per year
max_sales = np.max(sales_data[:,1:], axis = 0)
print("max_sale: ",max_sales)

#Average sale pe restaurant
average_sale = np.mean(sales_data[:,1:],axis = 1)
print("average_sale: ",average_sale)
```

```python
cumsum = np.cumsum(sales_data[:,1:], axis = 1)
print(cumsum)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ",vector1 + vector2)
print("\nMultiplication vector: ", vector1 * vector2)
print("\nDivision vector: ", vector1 / vector2)
print("\nDot Product: ", np.dot(vector1 , vector2))

angle = np.arccos(np.dot(vector1 , vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nAngle: ", np.degrees(angle)) #np.degrees to print angle in degree
#np.arccos computes the inverse cosine of a value in the range[-1,1]
#np.linalg.norm computes the magnitude(or Euclidean norm) of a vector
```
